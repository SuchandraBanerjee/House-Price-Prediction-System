#Plotting histogram

Histograms for better insights into the distribution of numerical data and also for comparison of feature distributions

```python
scaler = preprocessing.MinMaxScaler() #Creating instance of MinMaxScale which scales numeric features to specified range typically between 0 and 1
scaled_data = scaler.fit_transform(housing) #Applying fit_transform method of MinMaxScaler to Housing dataset to return the transformed data in scaled_data
num_features = min(len(housing.columns), 16)  # Limit the number of features to 16 to avoid overcrowded plots
num_rows = (num_features - 1) // 4 + 1  # Calculate the number of rows needed max 4 and min 1
num_cols = min(num_features, 4)  # Calculate the number of columns needed max 4
fig, axs = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(12, 3 * num_rows)) #returns array of axes and figure
fig.subplots_adjust(hspace=0.5, wspace=0.5)#adjusts space between subplots vertical and horizontal respectively
axs = axs.ravel()#flattens array of Axesto 1D array for easier itteration

for i, column in enumerate(housing.columns): #loop iterated over columns of housing dataframe
    if i >= num_features:
        break  # Exit the loop if all features have been plotted
    axs[i].hist(scaled_data[:, i], bins=30, color='skyblue', edgecolor='black')
    axs[i].set_title(column)
    axs[i].set_xlabel('Scaled Values')
    axs[i].set_ylabel(column+' Frequency')

# Remove any remaining empty plots
for j in range(num_features, num_rows * num_cols):
    fig.delaxes(axs[j])#iterating over the range of indices from num_features up to num_rows*num_cols which represents the indices of empty subplots

plt.show()#display the figure with the histograms
```

```python
housing = strat_train_set.drop("MEDV", axis=1)
housing_labels = strat_train_set["MEDV"].copy()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    #can add as many as needed
    ('std_scaler', StandardScaler()),
])
```

```python
housing_num_tr = my_pipeline.fit_transform(housing)
```

```python
housing_num_tr.shape
```

```python
housing_num_tr # a numpy array
```

#Model Training

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, VotingRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.svm import SVR

m1 = RandomForestRegressor()
m2 = DecisionTreeRegressor()
m3 = LinearRegression()
m4 = GradientBoostingRegressor()

# Fit each model separately
m1.fit(housing_num_tr, housing_labels)
m2.fit(housing_num_tr, housing_labels)
m3.fit(housing_num_tr, housing_labels)
m4.fit(housing_num_tr, housing_labels)


some_data = housing.iloc[:10]
some_labels = housing_labels.iloc[:10]
prepared_data = my_pipeline.transform(some_data)

# Make predictions for each model
m1_predictions = m1.predict(prepared_data)
m2_predictions = m2.predict(prepared_data)
m3_predictions = m3.predict(prepared_data)
m4_predictions = m4.predict(prepared_data)



# Fit the ensemble voting regressor on the 5 models
model = VotingRegressor(estimators=[('RF', m1), ('DTR', m2), ('LR', m3), ('GBR', m4)])
model.fit(housing_num_tr, housing_labels)

# Make predictions using the ensemble voting regressor
ensemble_predictions = model.predict(prepared_data)

# Store the results in a dictionary or DataFrame
results = {
    'RF': m1_predictions,
    'DTR': m2_predictions,
    'LR': m3_predictions,
    'GBR': m4_predictions,
    'Ensemble': ensemble_predictions
}

results_df = pd.DataFrame(results)
labels_list = list(some_labels)
results_df['Labels'] = labels_list

# Display the results
print(results_df)
```

#Evaluating the  model

```python
from sklearn.metrics import mean_squared_error

# Calculate RMSE for each model
housing_predictionsm1 = m1.predict(housing_num_tr)
housing_predictionsm2 = m2.predict(housing_num_tr)
housing_predictionsm3 = m3.predict(housing_num_tr)
housing_predictionsm4 = m4.predict(housing_num_tr)

m1_rmse = np.sqrt(mean_squared_error(housing_labels, housing_predictionsm1))
m2_rmse = np.sqrt(mean_squared_error(housing_labels, housing_predictionsm2))
m3_rmse = np.sqrt(mean_squared_error(housing_labels, housing_predictionsm3))
m4_rmse = np.sqrt(mean_squared_error(housing_labels, housing_predictionsm4))

# Set precision to 4 digits
np.set_printoptions(precision=4)

# Display the RMSE for each individual model
print("RMSE for Models")
print("Random Forest: ", format(m1_rmse, '.4f'))
print("Decision Tree: ", format(m2_rmse, '.4f'))
print("Linear Regression: ", format(m3_rmse, '.4f'))
print("Gradient Boosting: ", format(m4_rmse, '.4f'))

housing_predictions = model.predict(housing_num_tr)
mse = mean_squared_error(housing_labels, housing_predictions)
rmse = np.sqrt(mse)

# Display the ensemble MSE and RMSE with precision of 4 digits
print("\nEnsemble MSE: ", format(mse, '.4f'))
print("Ensemble RMSE: ", format(rmse, '.4f'))
```

#Using better evaluation, cross validation

```python
from sklearn.model_selection import cross_val_score
# Calculate RMSE scores
scores = cross_val_score(model, housing_num_tr, housing_labels,
                         scoring="neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-scores)

# Calculate mean and standard deviation
mean = np.mean(rmse_scores)
std_dev = np.std(rmse_scores)

# Set precision to 4
mean = round(mean, 4)
std_dev = round(std_dev, 4)

# Print the results
print("RMSE Scores:", rmse_scores.round(4))
print("Mean:", mean)
print("Standard Deviation:", std_dev)
```

#Model Testing  on Test Set

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import joblib

# Load the saved model
loaded_model = joblib.load('HPP.pkl')

# Prepare the test data
X_test = strat_test_set.drop("MEDV", axis=1)
Y_test = strat_test_set["MEDV"].copy()
X_test_prepared = my_pipeline.transform(X_test)

# Make predictions on the test data by loaded model
final_predictions = loaded_model.predict(X_test_prepared)


# Calculate the evaluation metrics
final_mse = mean_squared_error(Y_test, final_predictions)
final_rmse = np.sqrt(final_mse)
mae = mean_absolute_error(Y_test, final_predictions)
r2 = r2_score(Y_test, final_predictions)
# Calculate the mean percentage error
percentage_errors = (final_predictions - Y_test) / Y_test
mean_percentage_error = np.mean(percentage_errors) * 100



# Print the predictions and the actual values
print("\n Predictions:", final_predictions)

# Print the evaluation metrics
print("\nMean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", final_rmse)
print("R2 Score:", r2)
print("Mean Percentage Error:", mean_percentage_error)
```
